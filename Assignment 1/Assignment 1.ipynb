# NUMERICAL ANALYSIS ON DATASET

```python
df['fastestLapSpeed'] = pd.to_numeric(df['fastestLapSpeed'], errors='coerce')
```

```python
colum = "points"
Q1 = df[colum].quantile(0.25)
Q3 = df[colum].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df[colum] < (Q1 - 1.5 * IQR)) | (df[colum] > (Q3 + 1.5 * IQR))]
plt.boxplot(df[colum])
plt.title("Boxplot of Fastest Lap Speed")
plt.ylabel("Speed")
plt.show()
plt.scatter(x=outliers.index, y=outliers[colum],color='green')
```

```python
Q1 = df["points"].quantile(0.25)
Q3 = df["points"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df = df[(df['points'] >= lower) & (df['points'] <= upper)]
```

```python
colum = "points"
Q1 = df[colum].quantile(0.25)
Q3 = df[colum].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df[colum] < (Q1 - 1.5 * IQR)) | (df[colum] > (Q3 + 1.5 * IQR))]
plt.boxplot(df[colum])
plt.title("Boxplot of Fastest Lap Speed")
plt.ylabel("Speed")
plt.show()
plt.scatter(x=outliers.index, y=outliers[colum],color='green')
```

Now we will try to categorize the columns into Categorical,Numerical and Identifiers

```python
cat = []
num = []        
identifier = []
exclude = ['resultId','raceId','driverId','time','fastestLapTime','fastestLapSpeed','statusId','milliseconds']
for column in df.columns:
    if column in exclude:
        identifier.append(column)
    elif pd.api.types.is_numeric_dtype(df[column]):
        if df[column].nunique() > 10:
            num.append(column)
        else:
            cat.append(column)
    else:
        cat.append(column)  
print("Categorical columns:", cat)
print("Numerical columns:", num)
print("Identifier columns:", identifier)
    
```
